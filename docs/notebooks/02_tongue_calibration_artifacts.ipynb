# 02 Tongue Calibration Artifacts

This notebook shows the calibrated Tongue container and its main report surfaces.

Full workflow context: [Tongue Calibration](../workflows/tongue-calibration.md)

Artifacts read here: the calibration container, calibration report, and resolved restart metadata.

```python
from pathlib import Path

from swimrs.container import SwimContainer

ROOT = Path("/nas/swim/examples/tongue_ensemble")
CALIBRATION_CONTAINER = ROOT / "data" / "tongue_ensemble.swim"

container = SwimContainer.open(str(CALIBRATION_CONTAINER), mode="r")
CALIBRATION_CONTAINER
```

```python
report = container.calibration_report()
print(report.summary())
```

```python
report.to_json()
```

```python
params_df = report.to_dataframe()
params_df.sort_values("irr", ascending=False).head(10)
```
